### Importing the packages

```python
# Load the packages
import requests
from bs4 import BeautifulSoup
```

### Making a get request

```python
# Defining the url of the site
base_site = "https://en.wikipedia.org/wiki/Music"

# Making a get request
response = requests.get(base_site)
response
```

```python
# Extracting the HTML
html = response.content
```

### Making the soup

```python
# Convert HTML to a BeautifulSoup object. This will allow us to parse out content from the HTML more easily.
# Using the default parser as it is included in Python
soup = BeautifulSoup(html, "html.parser")
```

### 1. Extract all existing titles of links

```python
# Find all links on the page
# Basicall all the <a> tags
links = soup.find_all("a")
```

```python
# Dropping the links without 'href' attribute
# This will drop the links that are not actually links like <a> tags that are used for styling
links = [link for link in links if link.get("href") is not None]
```

```python
# Getting all titles
titles = [link.get("title") for link in links if link.get("title") is not None]
# Dropping the titles that are not actually titles
titles = [title for title in titles if title != "Wikipedia:WikiProject Music"]
```

```python
# Removing the 'None' titles
titles = [title for title in titles if title is not None]
# Removing the duplicates
titles = list(set(titles))
# Removing the empty titles
titles = [title for title in titles if title != ""]
```

### 2. Extract all heading 2 strings.

```python
# Inspect all h2 tags
h2_tags = soup.find_all("h2")
```

```python
# Get the text
h2_text = [h2.get_text() for h2 in h2_tags]
# Dropping the empty text
h2_text = [text for text in h2_text if text != ""]
```

### 3. Print the whole footer text.

```python
# By inspection: we see that the footer is contained inside a ...
# Dropping the footer
h2_text = [text for text in h2_text if text != "Contents"]
# Dropping the empty text
h2_text = [text for text in h2_text if text != "See also"]
# Dropping the empty text
h2_text = [text for text in h2_text if text != "References"]
# Dropping the empty text
h2_text = [text for text in h2_text if text != "External links"]
# Dropping the empty text
h2_text = [text for text in h2_text if text != "Further reading"]
# Dropping the empty text
h2_text = [text for text in h2_text if text != "Notes"]

#  Printing the titles
for title in titles:
    print(title)
    
```
